```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('matches.csv')
```

```python
data.head()
```

```python
data.head()
```

```python
data['player_of_match'].value_counts()
```

```python
data['player_of_match'].value_counts().head()
```

```python
list(data['player_of_match'].value_counts().head().keys())
list(data['player_of_match'].value_counts().head().values)
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(data['player_of_match'].value_counts().head().keys()), list(data['player_of_match'].value_counts().head().values))
plt.show()
```

```python
data['result'].value_counts()
```

```python
data['toss_winner'].value_counts()
```

```python
batting_first = data[(data['win_by_runs']!=0) & (data['toss_decision']=='bat')]
```

```python
batting_first
```

```python
plt.figure(figsize=(7,7))
plt.hist(batting_first['win_by_runs'])
plt.title("Distribution of runs")
plt.xlabel('Runs')
plt.ylabel('Wins')
plt.show()
```

```python
batting_first['win_by_runs'].max()
```

```python
batting_first['win_by_runs'].min()
```

```python
batting_first['winner'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.bar(batting_first['winner'].value_counts().keys(), batting_first['winner'].value_counts().values)
plt.title("Most Winning Teams")
plt.xlabel('Teams')
plt.ylabel('No. of Wins')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()), labels = list(batting_first['winner'].value_counts().keys()), autopct = '%0.1f%%',)
plt.title("Most Winning Teams")
plt.show()
```
